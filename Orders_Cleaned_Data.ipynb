```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
```

```python
file_path = '/content/drive/MyDrive/DataSets/orders.csv'
df = pd.read_csv(file_path)
```

```python
df.head(20)
```

```python
df["Ship Mode"].unique()
```

```python
df["Ship Mode"].replace(['Not Available','unknown'], np.nan, inplace = True)
```

```python
df.rename(columns=lambda x: x.lower().replace(' ', '_'), inplace=True)
df.columns
```

```python
df["discount_per_unit"] = df["list_price"] * (df["discount_percent"] / 100)
df["sale_price"] = (df["list_price"] - df["discount_per_unit"])*df["quantity"]
df["profit"] = df["sale_price"] - (df["cost_price"] * df["quantity"])
```

```python
df.head(20)
```

```python
df.columns
```

```python
df.drop(['discount_percent'], inplace = True, axis='columns')
```

```python
df.columns
```

```python
df.dtypes
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], format="%Y-%m-%d")
```
